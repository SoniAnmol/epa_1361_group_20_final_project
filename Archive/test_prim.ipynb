```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import scipy as sp
```

**without any policies:**

```python
# Reading the results file
from ema_workbench import load_results
file_name = 'test_zeropolicies.tar.gz'
results = load_results(file_name)
# Extracting experiment and outcomes from results
experiments, outcomes = results
```

```python
experiments, outcomes = results
policies = experiments['policy']

data = pd.DataFrame.from_dict(outcomes)
data['policy'] = policies

sns.pairplot(data, hue='policy',  vars=outcomes.keys(), )
plt.show()
```

```python
experiments
```

```python
experiments.loc[10]
```

```python
outcomes
```

```python
from ema_workbench.analysis import prim

data = outcomes['All Costs']

y = data > np.percentile(data, 80)

prim_alg = prim.Prim(experiments,y, threshold=0.8)
box1 = prim_alg.find_box()
```

```python
box1.show_tradeoff()
plt.show()
```

```python
box1.inspect_tradeoff()
```

```python
box1.inspect(23)
box1.inspect(23, style='graph')
plt.show()
```

```python
box1.select(23)
fig = box1.show_pairs_scatter()
fig.set_size_inches((12,12))
plt.show()
```

**With randomly generated policies:**


```python
# Reading the results file
from ema_workbench import load_results
file_name = 'test_policies.tar.gz'
results = load_results(file_name)
# Extracting experiment and outcomes from results
experiments_p, outcomes_p = results
```

```python
experiments, outcomes = results
policies = experiments['policy']

data = pd.DataFrame.from_dict(outcomes)
data['policy'] = policies

sns.pairplot(data, hue='policy',  vars=outcomes.keys(), )
plt.show()
```

```python
experiments_p.loc[0] 
```

```python
experiments_p.loc[100]
```

```python
outcomes_p
```


**costs:**

```python
from ema_workbench.analysis import prim
data = outcomes_p['All Costs']

y = data < np.percentile(data, 10)

prim_alg = prim.Prim(experiments_p,y, threshold=0.9, peel_alpha=0.03)
box1 = prim_alg.find_box()
```

```python
box1.inspect_tradeoff()
```

```python
box1.inspect(7)
box1.inspect(7, style='graph')
plt.show()
```

```python
box1.select(7)
fig = box1.show_pairs_scatter()
fig.set_size_inches((12,12))
plt.show()
```

**casualties:**

```python
from ema_workbench.analysis import prim
data = outcomes_p['Expected Number of Deaths']

y = data < np.percentile(data, 20)

prim_alg = prim.Prim(experiments_p, y, threshold=0.9)
box1 = prim_alg.find_box()
```

```python
box1.inspect_tradeoff()
```
