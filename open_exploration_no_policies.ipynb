```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import scipy as sp
```

```python
# Reading the results file
from ema_workbench import load_results
file_name = 'nopolicy.tar.gz'
results = load_results(file_name)
# Extracting experiment and outcomes from results
experiments, outcomes = results
```

```python
df_outcomes = pd.DataFrame.from_dict(outcomes)
```

```python
sns.pairplot(df_outcomes)
plt.show()
```

```python
df_outcomes.corr()
```

```python
from ema_workbench.analysis import prim

data = df_outcomes["Expected Annual Damage"]

y1 = data > np.percentile(data, 80)

prim_alg = prim.Prim(experiments,y1, threshold=0.8, peel_alpha=0.05)
box1 = prim_alg.find_box()
```

```python
box1.inspect_tradeoff()
```

```python
box1.inspect(24)
box1.inspect(24, style='graph')
plt.show()
```

```python
box1.select(24)
fig = box1.show_pairs_scatter()
fig.set_size_inches((12,12))
plt.show()
```

```python
from ema_workbench.analysis import dimensional_stacking
```

```python
dimensional_stacking.create_pivot_plot(experiments, y1)
plt.show()
```

lots of empty spaces and thus we need more experiments

**The most important aspects found from the prim analysis and the dimensional stacking is that the most damage occurs when dikes in area 1 fail**
